```python
# Part 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/college_completion.csv',low_memory=False)
```

```python
# Part 2 - The dataset contains 3,798 observations and 63 variables, including institution level, control type, average student aid, and graduation rates.
print( df.shape, '\n')
print( df.head(), '\n')
```

```python
# Part 3
# The data shows more 2-year public institutions and more 4-year private not-for-profit institutions. Private for-profit institutions are almost evenly split between 2-year and
# 4-year levels.
print(pd.crosstab(df['control'],df['level']),'\n')
print(pd.crosstab(df['control'],df['level'], normalize = 'all'),'\n')
```

```python
# Part 4
import seaborn as sns

# Histogram
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['grad_100_value'], kde=False)
plt.title('Histogram of grad_100_value')

# Kernel Density Plot
plt.subplot(2, 2, 2)
sns.kdeplot(df['grad_100_value'])
plt.title('Kernel Density Plot of grad_100_value')

# Boxplot
plt.subplot(2, 2, 3)
sns.boxplot(x=df['grad_100_value'])
plt.title('Boxplot of grad_100_value')

plt.tight_layout()
plt.show()

# Statistical description
stat_desc = df['grad_100_value'].describe()
stat_desc
```

```python
# Part 5
# Private not-for-profit institutions tend to have higher graduation rates than public and private for-profit institutions.
# 4-year institutions generally have higher graduation rates compared to 2-year institutions.
sns.kdeplot(data=df,x='grad_100_value',hue='control')
```

```python
sns.kdeplot(data=df,x='grad_100_value',hue='level')
```

```python
grouped_stat_desc_level = df.groupby('level')['grad_100_value'].describe()
grouped_stat_desc_control = df.groupby('control')['grad_100_value'].describe()

(grouped_stat_desc_level, grouped_stat_desc_control)
```

```python
# Part 6
# We can see that the private for-profit 2-year rate is almost double the private for-profit 4-year rate. There is little difference between
# 2-year and 4-year programs from private not-for-profit institutions. Lastly, public 2-year rates are the worst by far.
df.loc[:,['level','control','grad_100_value']].groupby(['level','control']).describe()
```

```python
df['levelXcontrol']=df['level']+', '+df['control']
sns.kdeplot(data=df,x='grad_100_value',hue='levelXcontrol')
```

```python
# Part 7
# The kernel density plot of aid_value reveals a wide range of aid values, with a distinct peak for lower values. Grouping by level and control
# shows that 4-year, private not-for-profit institutions tend to offer the highest average aid, which likely contributes to their higher graduation rates.
df['aid_arcsinh'] = np.arcsinh(df['aid_value'])

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.kdeplot(data=df, x='aid_arcsinh')
plt.title('Kernel Density of aid_arcsinh')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='aid_value')
plt.title('Kernel Density of aid_value')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.kdeplot(data=df, x='aid_value', hue='control')
plt.title('Kernel Density of aid_value by Control')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='aid_value', hue='level')
plt.title('Kernel Density of aid_value by Level')

plt.tight_layout()
plt.show()
```

```python
df.loc[:,['control','aid_value']].groupby('control').describe()
```

```python
df.loc[:,['level','aid_value']].groupby('level').describe()
```

```python
# Part 8
# The scatterplot shows a positive correlation between aid_value and grad_100_value, indicating that higher student aid is associated with higher graduation rates.
# Grouping by level and control in the scatterplot highlights that 4-year, private not-for-profit institutions not only offer higher aid but also tend to have higher
# graduation rates, suggesting a stronger relationship between aid and graduation rates in these institutions.
sns.scatterplot(data=df,x='aid_value',y='grad_100_value')
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='control')
plt.title('Scatter Plot of grad_100_value by aid_value (Grouped by Control)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='level')
plt.title('Scatter Plot of grad_100_value by aid_value (Grouped by Level)')

plt.tight_layout()
plt.show()
```

```python
# Part 2
# The data dictionary for the Medicaid Spending by Drug dataset provides detailed definitions for each variable, including brand name, generic name, total spending for various years,
# It does not explicitly mention how missing values are handled or describe possible values and missing codes for categorical variables, nor does it detail missing value
# treatment for numeric variables. Overall, the data dictionary offers clear variable definitions and methodology for things like outlier identification but lacks specifics on
# handling missing data.
```

```python
# Part 3
# Brnd_Name and Gnrc_Name: These categorical variables give insight into the variety of drugs in terms of brand and generic names, which is interesting for analyzing spending
# patterns across different types of medications.
# Tot_Spndng_2020 and Tot_Spndng_2021: This numeric variable is interesting because it shows the financial impact on Medicaid.
# Tot_Dsg_Unts_2020: This numeric variable is interesting because it offers insight into the volume of distribution and utilization of each drug.

import pandas as pd

data = pd.read_csv('./data/DSD_MCD_RY22_P06_V20_D21_BGM.csv',low_memory=False)

# The selected variables
missing_values = data[['Gnrc_Name', 'Brnd_Name', 'Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020']].isnull().sum()
print(missing_values)

# Standardizing text data for 'Gnrc_Name' and 'Brnd_Name'
data['Gnrc_Name'] = data['Gnrc_Name'].str.strip().str.upper()
data['Brnd_Name'] = data['Brnd_Name'].str.strip().str.upper()

# Ensuring numerical data integrity for 'Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Spndng_2020', and 'Tot_Dsg_Unts_2020'
# Filling missing values with 0 and ensuring they are non-negative
numerical_columns = ['Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020']
data[numerical_columns] = data[numerical_columns].fillna(0).clip(lower=0)

# missing_values = data[['Gnrc_Name', 'Brnd_Name', 'Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020']].isnull().sum()
# print(missing_values)
```

```python
# Part 4
# Yes there were outliers, but the limits of the x-axis were set to better see the rest of the data.
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

numeric_data = data[['Tot_Spndng_2021', 'Tot_Dsg_Unts_2021', 'Tot_Spndng_2020', 'Tot_Dsg_Unts_2020']]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Boxplots for Numeric Variables', fontsize=16)

sns.boxplot(ax=axes[0, 0], data=numeric_data, x='Tot_Spndng_2021')
axes[0, 0].set_title('Total Spending 2021')
axes[0, 0].set_xlim(numeric_data['Tot_Spndng_2021'].min(), numeric_data['Tot_Spndng_2021'].quantile(0.95))

sns.boxplot(ax=axes[0, 1], data=numeric_data, x='Tot_Dsg_Unts_2021')
axes[0, 1].set_title('Total Dosage Units 2021')
axes[0, 1].set_xlim(numeric_data['Tot_Dsg_Unts_2021'].min(), numeric_data['Tot_Dsg_Unts_2021'].quantile(0.95))

sns.boxplot(ax=axes[1, 0], data=numeric_data, x='Tot_Spndng_2020')
axes[1, 0].set_title('Total Spending 2020')
axes[1, 0].set_xlim(numeric_data['Tot_Spndng_2020'].min(), numeric_data['Tot_Spndng_2020'].quantile(0.95))

sns.boxplot(ax=axes[1, 1], data=numeric_data, x='Tot_Dsg_Unts_2020')
axes[1, 1].set_title('Total Dosage Units 2020')
axes[1, 1].set_xlim(numeric_data['Tot_Dsg_Unts_2020'].min(), numeric_data['Tot_Dsg_Unts_2020'].quantile(0.95))

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
top_generic_names = data['Gnrc_Name'].value_counts().head(10)

plt.figure(figsize=(16, 6))
sns.barplot(x=top_generic_names.index, y=top_generic_names.values)
plt.title('Top 10 Generic Names')
plt.xticks(rotation=90)
plt.show()

top_brand_names = data['Brnd_Name'].value_counts().head(10)

plt.figure(figsize=(16, 6))
sns.barplot(x=top_brand_names.index, y=top_brand_names.values)
plt.title('Top 10 Brand Names')
plt.xticks(rotation=90)
plt.show()
```

```python
# Part 5
# The first graph shows that the distribution is highly skewedto the right. The scale of the x-axis and concentration of data near zero might mean there are some extreme outliers affecting the distribution.
# After applying a log transformation, the distribution looks a lot more like a bell curve.

spending_2021 = data['Tot_Spndng_2021']


sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.kdeplot(spending_2021, bw_adjust=0.5)
plt.title('KDE of Total Spending 2021')
plt.xlabel('Total Spending 2021')
plt.ylabel('Density')
plt.show()


spending_2021_log = np.log1p(spending_2021)

plt.figure(figsize=(12, 6))
sns.kdeplot(spending_2021_log, bw_adjust=0.5)
plt.title('KDE of Log-Transformed Total Spending 2021')
plt.xlabel('Log-Transformed Total Spending 2021')
plt.ylabel('Frequency')
plt.show()
```

```python
# Part 6
# A log transformation was applied to handle the wide range of values and to get a better look at the data. We observe a positive correlation
# between total spending and total dosage units, which makes sense as it would cost more to dispense more dosages of medication.



data['Log_Total_Spending_2021'] = np.log1p(data['Tot_Spndng_2021'])
data['Log_Total_Dosage_Units_2021'] = np.log1p(data['Tot_Dsg_Unts_2021'])

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log_Total_Spending_2021', y='Log_Total_Dosage_Units_2021', data=data)

plt.xlim(left=0, right=25)
plt.ylim(bottom=0, top=25)

plt.title('Scatter Plot of Log-Transformed Total Spending vs. Total Dosage Units 2021')
plt.xlabel('Log of Total Spending 2021')
plt.ylabel('Log of Total Dosage Units 2021')
plt.show()

```
